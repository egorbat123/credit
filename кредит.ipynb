Пусть мы имеем следующие данные:
*   summ - Сумма кредита (руб.)
*   period - Срок, на который берётся кредит (кол-во лет)
*   perc - Процентная ставка (в десятичном выражении, например 0.15)

Рассчитаем:
*   mp - Ежемесячный взнос (руб.)
*   total - Общую сумму выплат (руб.)

Допущения:
*   Схема кредита - аннуитетная (ежемесячный платёж всегда одинаковый)
*   Досрочное погашение отсутвует
*   Платеж производится в конце периода
*   Проценты и налоги могут быть опущены

```python
def interest(period, perc, summ):
    mp_cnt = period * 12       # сколько всего месяцев
    r = perc / 12.0            # процентная ставка за один месяц
    ak = (r * (1 + r) ** mp_cnt) / (((1 + r) ** mp_cnt) - 1) # аннуитетный коэффициент
    mp = summ * ak             # размер ежемесячного платежа
    total = mp * mp_cnt        # общая сумма выплат
    return round(mp, 2), round(total, 2)
```

```python
summ = 1000000
period = 10
perc = 0.15

mp, total = interest(period, perc, summ)
print('Ежемесячный платёж = ', mp, ',  Общая сумма выплат = ', total)
```
